# Predictive Maintenance ML Project

This notebook contains data loading, EDA, and setup for the AI4I 2020 Predictive Maintenance dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset
Upload `ai4i2020.csv` into the Colab environment then run:

```
df = pd.read_csv('ai4i2020.csv')
df.head()
```

## Basic Info

```
df.shape, df.info(), df.describe()
```

## Missing Values

```
df.isnull().sum()
```

## Failure Label Distribution

```
df['Machine failure'].value_counts()
```

## Pairplot

```
sns.pairplot(df, hue='Machine failure')
```

## Correlation Heatmap

```
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
